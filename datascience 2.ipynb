```python
homeless=pd.read_csv('C:/Users/goswami/Jupyter Notebook/homelessness.csv')
```

```python
homeless
```

```python
homeless=homeless.drop(columns='Unnamed: 0')
```

```python
homeless_reg=homeless.groupby('region').agg({'individuals':'sum'}).sort_values('individuals',ascending=False)
```

```python
plt.barh(homeless_reg.index,homeless_reg.individuals,color='#0a9396')

plt.xlabel('homeless count')
plt.show()
```
